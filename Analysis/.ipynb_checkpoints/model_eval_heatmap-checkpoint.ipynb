```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
game_rst = pd.read_csv(os.getcwd() + '/' + 'game_rst.txt', sep = '\t')
```

```python
def getModelMap(data):
#     game_rst = pd.read_csv(os.getcwd() + '/' + 'game_rst.txt', sep = '\t')
    model_name = list(set(game_rst['model_1']))
    model_map = {m: {m: 0 for m in model_name} for m in model_name}
    for i in range(game_rst.shape[0]): #300
        model_map[game_rst['model_1'][i]][game_rst['model_2'][i]] += 1
    return model_map, model_name
```

```python
def getModelTable(data):
    model_map, model_name = getModelMap(data)
    model_matrix = [np.array(list(m.values())) for m in model_map.values()]
    model_table = pd.DataFrame(model_matrix, index = [m_name for m_name in model_name], 
                      columns = model_name) # index: model_1, columns: model_2
    return model_table
```

```python
getModelTable(game_rst)
```

```python
model_table= getModelTable(game_rst)
```

```python
model_table.sum() # number of games each model plays
```

```python
def plot(data):
    model_table = getModelTable(data)
    sns.heatmap(model_table, cmap="RdBu_r")
```

```python
plot(game_rst)
```
